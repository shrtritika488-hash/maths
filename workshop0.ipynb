```python
# 6
def sum_nested_list(lst):
    total = 0
    for x in lst:
        if isinstance(x, list):
            total += sum_nested_list(x)
        else:
            total += x
    return total

sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]])
```

```python
# 7
def generate_permutations(s):
    if len(s) <= 1:
        return [s]
    out = []
    for i in range(len(s)):
        for p in generate_permutations(s[:i] + s[i+1:]):
            if s[i] + p not in out:
                out.append(s[i] + p)
    return out

generate_permutations("abc")
```

```python
# 8
def calculate_directory_size(d):
    total = 0
    for k, v in d.items():
        if isinstance(v, dict):
            total += calculate_directory_size(v)
        else:
            total += v
    return total

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {"file5.txt": 250},
        "file6.txt": 150
    }
}

calculate_directory_size(directory_structure)
```

```python
# 9
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

min_coins([1, 2, 5], 11)
```

```python
# 10
def longest_common_subsequence(s1, s2):
    n, m = len(s1), len(s2)
    dp = [[0]*(m+1) for _ in range(n+1)]

    for i in range(1, n+1):
        for j in range(1, m+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[n][m]

longest_common_subsequence("abcde", "ace")
```

```python
# 11
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0]*(capacity+1) for _ in range(n+1)]

    for i in range(1, n+1):
        w = weights[i-1]
        v = values[i-1]
        for c in range(1, capacity+1):
            if w <= c:
                dp[i][c] = max(dp[i-1][c], dp[i-1][c-w] + v)
            else:
                dp[i][c] = dp[i-1][c]
    return dp[n][capacity]

knapsack([1, 3, 4, 5], [1, 4, 5, 7], 7)
```
