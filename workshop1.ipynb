```python
import numpy as np

#1.Array from 10 to 49
a = np.arange(10, 50)
print(a)

#2. 3x3 matrix from 0 to 8
b = np.arange(9).reshape(3,3)
print(b)

#3. Identity matrix 3x3
c = np.eye(3)
print(c)

#4. Random array of size 30 + mean
d = np.random.random(30)
print("Mean:", d.mean())

#5. 10x10 random array → min & max
e = np.random.random((10,10))
print("Min:", e.min(), "Max:", e.max())

#6. Zero array size 10 → replace 5th with 1
f = np.zeros(10)
f[4] = 1     # index starts at 0
print(f)

#7. Reverse array
arr = np.array([1,2,0,0,4,0])
g = arr[::-1]
print(g)

#8. 2D array with 1 on border, 0 inside
h = np.ones((5,5))
h[1:-1, 1:-1] = 0
print(h)

#9. 8x8 checkerboard pattern
i = np.zeros((8,8))
i[1::2, ::2] = 1
i[::2, 1::2] = 1
print(i)
```

```python
import numpy as np

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

#1 Addition
print(x + y)

#2 Subtraction
print(x - y)

#3 Multiply with integer
print(x * 3)

#4 Square of each element
print(x**2)

#5 Dot products
print("v·w =", np.dot(v,w))
print("x·v =", np.dot(x,v))
print("x·y =", np.dot(x,y))

#6 Concatenate x & y (row) and v & w (column)
print(np.concatenate((x,y), axis=0))
print(np.vstack((v,w)).T)

#7 Concatenate x & v → explanation
try:
    print(np.concatenate((x,v)))
except Exception as e:
    print("Error:", e)
    print("Reason: x is 2D but v is 1D → shapes do not match!")
```

```python
import numpy as np

#------------------------------
# Problem 4 – Matrix Operations
#------------------------------
A = np.array([[3,4],
              [7,8]])

B = np.array([[5,3],
              [2,1]])

#1 Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
print("A * A^-1 =\n", A @ A_inv)

#2 Prove AB != BA
AB = A @ B
BA = B @ A
print("\nA * B =\n", AB)
print("\nB * A =\n", BA)

#3 Prove (AB)^T = B^T * A^T
print("\n(AB)^T =\n", AB.T)
print("\nB^T * A^T =\n", B.T @ A.T)

#------------------------------
# Solve Linear Equations
#------------------------------
# System:
# 2x – 3y +  z = –1
# x –  y + 2z = –3
# 3x +  y –  z = 9
A = np.array([[ 2, -3, 1],
              [ 1, -1, 2],
              [ 3,  1, -1]])

B = np.array([-1, -3, 9])

# Solve using linalg.solve
solution = np.linalg.solve(A, B)
print("\nSolution using solve() =\n", solution)

# Solve using inverse
A_inv = np.linalg.inv(A)
solution2 = A_inv @ B
print("\nSolution using inverse =\n", solution2)
```
